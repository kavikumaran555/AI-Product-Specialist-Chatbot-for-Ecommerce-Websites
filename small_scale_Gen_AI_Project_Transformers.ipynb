```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Load model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Product dataset
data = {
"product":[
"wireless mouse",
"gaming keyboard",
"bluetooth headphones",
"power bank",
"smart watch",
"laptop stand",
"usb c hub",
"external hard drive",
"gaming mouse",
"monitor",
"mechanical keyboard",
"gaming headset",
"tablet",
"laptop",
"smartphone",
"webcam",
"microphone",
"portable speaker",
"ssd drive",
"router",
"smart tv",
"led tv",
"refrigerator",
"washing machine",
"air conditioner",
"microwave oven",
"electric kettle",
"coffee maker",
"blender",
"induction stove",
"printer",
"scanner",
"digital camera",
"dslr camera",
"tripod",
"graphics tablet",
"vr headset",
"projector",
"desk lamp",
"office chair",
"tv"
],

"description":[
"Wireless mouse connects using bluetooth or USB receiver with long battery life.",
"Gaming keyboard has mechanical switches and RGB lighting for gaming.",
"Bluetooth headphones provide wireless audio and noise cancellation.",
"Power bank is a portable charger used to charge phones and tablets.",
"Smart watch tracks fitness, heart rate, steps and notifications.",
"Laptop stand improves posture and cooling for laptops.",
"USB C hub expands ports with HDMI USB and SD card slots.",
"External hard drive stores large files and backups.",
"Gaming mouse offers high DPI sensor and programmable buttons.",
"Full HD monitor provides clear display for work and gaming.",
"Mechanical keyboard provides tactile feedback and durability.",
"Gaming headset delivers surround sound and microphone support.",
"Tablet is a portable touchscreen device for entertainment and work.",
"Laptop is a portable computer used for work study and browsing.",
"Smartphone is a mobile phone with apps internet and camera.",
"Webcam is used for video calls and online meetings.",
"Microphone captures clear audio for streaming and recording.",
"Portable speaker plays wireless music using bluetooth.",
"SSD drive provides fast storage for computers and laptops.",
"Router provides wireless internet connectivity in homes and offices.",
"Smart TV connects to the internet and streams content from apps like Netflix and YouTube.",
"LED TV provides bright display with energy efficient LED technology.",
"Refrigerator keeps food fresh using cooling and freezing compartments.",
"Washing machine automatically washes clothes with multiple wash programs.",
"Air conditioner cools rooms and controls indoor temperature.",
"Microwave oven heats and cooks food quickly using microwave energy.",
"Electric kettle quickly boils water for tea coffee or cooking.",
"Coffee maker brews fresh coffee automatically.",
"Blender mixes fruits vegetables and ingredients for smoothies and juices.",
"Induction stove cooks food using electromagnetic heating technology.",
"Printer prints documents and photos from computers and phones.",
"Scanner converts paper documents into digital files.",
"Digital camera captures high quality photos and videos.",
"DSLR camera provides professional photography with interchangeable lenses.",
"Tripod stabilizes cameras for photography and video recording.",
"Graphics tablet allows digital drawing and design on computers.",
"VR headset provides immersive virtual reality gaming experience.",
"Projector displays large screen images for movies and presentations.",
"Desk lamp provides focused lighting for study or work.",
"Office chair provides ergonomic support for long working hours.",
"Sure, would you like to go for LED TV or LCD TV?"
]
}

df = pd.DataFrame(data)

combined = df["product"] + " " + df["description"]
product_embeddings = model.encode(combined.tolist())

greetings = ["hi","hello","hey","are you there","anyone"]
courtesy = ["okay","thank you","thanks","i will buy","fine","alright"]
bye_words = ["bye","goodbye","see you","thats all","thats all thanks","thats it"]
how_are_you_phrases = ["how are you","how its going","how is it going","how are you doing","how is your day","how have you been","hows everything","everything good"]
# First bot message
chat_history = "<div style='text-align:left;margin:8px;color:darkblue'><b>Joseph AI Chat:</b> Hello! I am your AI product specialist. Please ask me about any product.</div>"

# Chat box
chat_area = widgets.HTML(
value=f"<div style='height:420px;width:520px;border:1px solid gray;padding:10px;overflow:auto;background:#f9f9f9'>{chat_history}</div>"
)

# Input
input_box = widgets.Text(
placeholder="Ask about a product..."
)

send_button = widgets.Button(description="Send")


def chatbot_response(message):

    user_msg = message.lower()

    if "like" in user_msg:
        return "Great!"

    # Convert query to embedding
    query_embedding = model.encode([user_msg])

    similarities = cosine_similarity(query_embedding, product_embeddings)

    best_match = np.argmax(similarities)

    score = similarities[0][best_match]

    # If product match is good → return product
    if score >= 0.20:

        product = df.iloc[best_match]["product"]
        description = df.iloc[best_match]["description"]

        return f"{product.title()}: {description}"

    # If no product match → check greetings
    if any(word in user_msg for word in greetings):
        return "Hello! Hope you have a great day. Happy to answer your questions"

    if any(word in user_msg for word in how_are_you_phrases):
        return "I'm doing great! How can I help you today?"

    if any(word in user_msg for word in courtesy):
        return "Great! I am here if you have anymore questions"

    if any(word in user_msg for word in bye_words):
        return "Its a pleasure to assist you. Goodbye! Have a nice day."

    return "Sorry, I do not have that product information available yet."


def send_message(b):

    global chat_history

    user_text = input_box.value

    if user_text == "":
        return

    bot_reply = chatbot_response(user_text)

    chat_history += f"<div style='text-align:right;margin:8px'><b>You:</b> {user_text}</div>"
    chat_history += f"<div style='text-align:left;margin:8px;color:darkblue'><b>Joseph AI:</b> {bot_reply}</div>"

    chat_area.value = f"<div style='height:420px;width:520px;border:1px solid gray;border-radius:10px;padding:10px;overflow:auto;background:#f9f9f9'>{chat_history}</div>"

    input_box.value = ""


send_button.on_click(send_message)

# allow enter key
input_box.on_submit(send_message)

display(chat_area, input_box, send_button)
```
